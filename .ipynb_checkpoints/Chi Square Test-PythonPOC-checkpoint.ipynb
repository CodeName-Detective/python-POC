```python
import pandas as pd
microsoft_data_df = pd.read_csv('Microsoft_Data.csv')

microsoft_data_df
```

```python
microsoft_data_df['OsBuild'] = microsoft_data_df['OsBuild'].astype('object') #Converted the column to the Object dtype

microsoft_data_df['OsBuild']
```

```python
#Doing the Chisquare test between OSBuild and Has Detection

import scipy.stats

crosstab = pd.crosstab(microsoft_data_df['OsBuild'],microsoft_data_df['HasDetections']) #Builds Coningency table or Cross tab or observed values table

scipy.stats.chi2_contingency(crosstab) #Outputs the (chi2 test statistic, p-value, degrees of freedom and expected values table)in the respective order

# As P-value < Alpha(0.05) we reject H0 i.e., Correlation Exsist 
```
