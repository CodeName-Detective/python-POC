```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random

#np.random.seed(43) # TO generate a the same values for different iterations

plt.figure(figsize=(20,5))

population =pd.DataFrame() #Create Empty Data frame

population['numbers'] = np.random.uniform(0,1000,size=10000) # Create an colum that is 1000 random numbers drawn from a uniform distribution

plt.hist(population['numbers'])

plt.show() #Population Distirbution
```

```python
sampled_means=[]

for i in range(0,1000):
    sampled_means.append(population['numbers'].sample(n=2000).mean())

plt.hist(pd.Series(sampled_means))

plt.show()

pd.Series(sampled_means).hist(bins=100)
```
