```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

df = pd.read_csv('Houseprices Dataset.csv')

plt.hist(df['SalePrice'])

plt.show()
```

```python
sampled_means = []

for i in range(0,1000):
    sampled_means.append(df['SalePrice'].sample(n=100).mean())

plt.hist(pd.Series(sampled_means))

plt.show()

pd.Series(sampled_means).hist(bins=100)
```
