```python
#Exercise 1-------------Get all columns whose data type is OBject?
import pandas as pd

house_price_df = pd.read_csv('Houseprices Dataset.csv')

all_columns = house_price_df.columns

obj_columns = []

for column in all_columns:
    if house_price_df[column].dtype == 'O':
        obj_columns.append(column)
obj_columns
```

```python
# Create A Data Frame with coulmn Name, Column Missing values count and their Percentage
import pandas as pd

house_price_df = pd.read_csv('Houseprices Dataset.csv')

missing_values_df =  pd.DataFrame({'ColumnName':house_price_df.columns,'MissingVlauesCount':house_price_df.isnull().sum(), 'MissingValuesPercentage': (house_price_df.isnull().sum()/house_price_df.shape[0])*100})

missing_values_df.sort_values(by='MissingValuesPercentage',ascending=False)
```
